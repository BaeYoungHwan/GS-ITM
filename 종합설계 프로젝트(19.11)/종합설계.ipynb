```python
from bs4 import BeautifulSoup
import re
import time
import requests
import json
from html import unescape
from selenium import webdriver
import time
```

```python
def download(method, url, param=None, data=None, timeout=1 , maxretires=3):
    try : 
        resp = requests.request(method, url, params=param, data=data, headers=headers)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e :
        if 500 <= e.response.status_code < 600 and maxretires > 0:
            print(maxretires)
            time.sleep(timeout)
            resp = download(method, url, param, data, timeout, maxretires-1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```

```python
def Craw(Data):
    driver = webdriver.Chrome()

    driver.get("https://www.naver.com")
    inputName = driver.find_element_by_xpath('//*[@id="query"]')
    inputName.send_keys(Data)
    driver.find_element_by_xpath('//*[@id="search_btn"]').click()
    
    Data1 = driver.find_element_by_xpath('//*[@id="nx_related_keywords"]/dl').text
    
    Data_blog = str()
    for x in driver.find_elements_by_class_name("thmb_bd"):
        x.click()
        time.sleep(1)
        driver.switch_to_window(driver.window_handles[-1])
        try:
            iframe = driver.find_element_by_name("mainFrame")
            driver.switch_to_frame(iframe)0
            Data_blog += driver.find_element_by_id("body").text
            
        except:
            pass
        driver.close()
        driver.switch_to_window(driver.window_handles[-1])
    driver.close()
    f = open(f'{Data}.text', 'w', encoding = 'UTF-8')
    f.write(Data1)
    f.write(Data_blog)
    f.close()
    
    return print(Data1, Data_blog)
```
